```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

# Cwiczenia
** Odtwórz poniższe wykresy, używając powyzszej ramki danych. Jest bardzo mało wskazówek, ponieważ większość wykresów można wykonać za pomocą zaledwie jednej lub dwóch linii kodu, a podpowiedź w zasadzie podałaby rozwiązanie. Zwróć szczególną uwagę na etykiety x i y.**


```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
# sns.set_style('whitegrid')
# titanic = sns.load_dataset('titanic')

sns.jointplot(x='fare',y='age',data=titanic)
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
# sns.set_style('whitegrid')
# titanic = sns.load_dataset('titanic')

sns.distplot(titanic['fare'],kde=False, bins=30, color = 'r')
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
# sns.set_style('whitegrid')
# titanic = sns.load_dataset('titanic')

sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
# sns.set_style('whitegrid')
# titanic = sns.load_dataset('titanic')

sns.countplot(x='sex',data=titanic)
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
# sns.set_style('whitegrid')
# titanic = sns.load_dataset('titanic')

heat_map = titanic.corr()
sns.heatmap(heat_map,cmap='coolwarm')
plt.xticks(rotation=0) # obracamy podpisy osi X poziomo!!!!!!!!!!!!!
plt.title('titanic.corr()')
```

```python
chart = sns.FacetGrid(data=titanic, col='sex') # z opcja: titatnic.dropna() wykres wyglada inaczej!
chart.map(sns.distplot, 'age',kde=False, color='navy')
# poprawic kolory i zakres osi Y
```
