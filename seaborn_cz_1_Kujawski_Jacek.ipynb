W tych ćwiczeniach będziemy pracować ze słynnym tytanicznym zestawem danych. Mowi on o tym, kto na Titanicu ma szanse przezyc a kto niestety nie.

### Wgranie danych

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Cwiczenia
** Odtwórz poniższe wykresy, używając powyzszej ramki danych. Jest bardzo mało wskazówek, ponieważ większość wykresów można wykonać za pomocą zaledwie jednej lub dwóch linii kodu, a podpowiedź w zasadzie podałaby rozwiązanie. Zwróć szczególną uwagę na etykiety x i y.**


```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic["fare"],kde=False, color = 'r')
```

```python
sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
tc = titanic.corr()
sns.heatmap(tc,cmap='coolwarm')
plt.title('titanic.corr()')
#poprawic podpisy!!!
```

```python
chart = sns.FacetGrid(data=titanic, col='sex')
chart.map(sns.distplot, 'age',kde=False, color = 'b')
# poprawic kolory i zakres osi Y
```
