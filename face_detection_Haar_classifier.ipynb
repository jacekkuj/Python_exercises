```python
import cv2
import matplotlib.pyplot as plt

# Picture import (i.e., 'face.jpg')
# https://petapixel.com/2019/02/19/this-website-generates-ai-portraits-of-people-who-dont-exist/
img = cv2.imread("C://Users//user//Documents//Dydaktyka//Analiza_Danych_w_Pythonie//non_existing_person.jpg") # your path
if img is None:
    raise FileNotFoundError("The image was not found! Please check the file path.")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Haar classifier works only for gray-scale

# Haar classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Face detection (change scaleFactor, minNeighbors)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Rectangle around detected face
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the result (matplotlib uses RGB colors, not BGR)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Detected faces')
plt.show()
```
