Narysuj pie chart, ktry bdzie podawa wartoci procentowej kadego pola. Nazwij kade pole dowolnie. Dodaj legend oraz spraw, zeby najmniejsze elementy byy odsunite od gwnego 'ciastka' (funkcja: explode)

```python
#import matplotlib.pyplot as plt
#import numpy as np
labels = 'A', 'B', 'C', 'E', 'F', 'G', 'H', 'I'
sizes = [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]
fig1, ax1 = plt.subplots()
explode = (0, 0, 0, 0.2, 0, 0, 0, 0)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%.2f %%', shadow=True, startangle=90, pctdistance=0.8)
#ax1.axis('equal')
plt.figure(figsize=(8,5))
plt.show()
```

## Subfigures

Mona utworzy figur z kilkoma podfigurami mozna uzyc polecenia `plt.subplots`. Tworzy siatk figur podrzdnych, w ktrej jako parametry podaje si liczb wierszy i kolumn w siatce. Zwraca par obiektw figury i tablic zawierajc podfigury. W matplotlib figury podrzdne nazywane s osiami. Zwr uwag na jedn literow rnic: o jest w liczbie pojedynczej, a osie w liczbie mnogiej od osi. Mona wic myle o osiach jako parze osi x i osi y, ktra definiuje dwuwymiarow (pod)figur. Przykad tego:

```python
fig, ax = plt.subplots(2,2)
print(ax.shape)
ax[0,0].plot(np.arange(6))          # top left
ax[0,1].plot(np.arange(6,0,-1))     # top right
ax[1,0].plot((-1)**np.arange(6))    # bottom left
ax[1,1].plot((-1)**np.arange(1,7))  # bottom right
plt.show()
```

Inny sposb wykonania powyzszego:

```python
plt.subplot(2, 2, 1)    
plt.plot(np.arange(6))
plt.subplot(2, 2, 2)
plt.plot(np.arange(6, 0, -1))
plt.subplot(2, 2, 3)
plt.plot((-1)**np.arange(6))
plt.subplot(2, 2, 4)
plt.plot((-1)**np.arange(1, 7))
plt.show()
```

Narysuj piechart. Podej wartosci przy kazdym kawalku ciasta.

```python
#import matplotlib.pyplot as plt
#import numpy as np
labels = 'United States', 'Great Britain', 'China', 'Russia', 'Germany'
sizes = [46, 27, 26, 19, 17]
fig1, ax1 = plt.subplots()
explode = (0, 0, 0.2, 0, 0)
ax1.pie(sizes, explode=explode, labels=labels, shadow=True, startangle=90, pctdistance=0.8)
ax1.axis('equal')
#plt.figure(figsize=(10,7), dpi=100)
#plt.style.use('ggplot')
plt.title('Country - gold medal')
plt.legend()
plt.show()
```
