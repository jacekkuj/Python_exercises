```python
zoo = pd.DataFrame([['elephant',1001,500],
['elephant',1002,600],
['elephant',1003,550],
['tiger',1004,300],
['tiger',1005,320],
['tiger',1006,330],
['tiger',1007,290],
['tiger',1008,310],
['zebra',1009,200],
['zebra',1010,220],
['zebra',1011,240],
['zebra',1012,230],
['zebra',1013,220],
['zebra',1014,100],
['zebra',1015,80],
['lion',1016,420],
['lion',1017,600],
['lion',1018,500],
['lion',1019,390],
['kangaroo',1020,410],
['kangaroo',1021,430],
['kangaroo',1022,410]], columns = ['animal','uniq_id','water_need'])
```

```python
zoo_eats = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], 
                         ['kangaroo','vegetables'], ['zebra','vegetables'], [
                             'giraffe','vegetables']], columns=['animal', 'food'])
```

```python
zoo_eats.head()
```

### Zad.1. Połącz powyższe tabele po kluczu 'animal' na nastepujace sposoby:
<br> - do klucza z prawej (SQL right outer join)
<br> - inner join

Powyzsze zadanie sprobuj wykonac za pomoca funkcji merge oraz join. 

```python
df_1 = pd.merge(zoo, zoo_eats, on = 'animal', how = 'inner')
df_1
```

```python
df_2 = pd.merge(zoo, zoo_eats, on = 'animal', how = 'right')
df_2
```

### dataset 2

```python
df1 = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
```

```python
df2 = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

### Zad.2. Sprobuj je polaczyc po podwojnym kluczu ['id','subject_id']. Ile rekordów zostanie? 

```python
df_3 = pd.merge(df1, df2, on = ['id', 'subject_id'])
df_3
```

### Zad.3. Połącz powyższe dane jako inner join z kluzem 'subject_id'

```python
df_4 = pd.merge(df1, df2, how = 'inner', on = ['subject_id'])
df_4
```
