```python
"""
This code loads a CSV file named big_five_dataset_08.csv into a Pandas DataFrame. 
It then selects the columns from EXT1 to EXT10 
and converts their values to the float type to ensure accurate calculations. 
In the next step, it sums the values of these ten columns 
for each row and stores the result in a new column named sum_EXT. 
This new column is inserted directly after the CON10 column in the data structure. 
Finally, the code saves the updated DataFrame 
to a new CSV file named big_five_dataset_08_with_sum_EXT.csv.
"""

import pandas as pd

# Path
df = pd.read_csv("C:\\Users\\user\\Documents\\Dydaktyka\\Analiza_Danych_w_Pythonie\\big_five_dataset_08.csv")
df
```

```python
# Change into float
ext_cols = [f'EXT{i}' for i in range(1, 11)]
df[ext_cols] = df[ext_cols].apply(pd.to_numeric, errors='coerce')
df[ext_cols]
```

```python
# Sum of EXT
df['sum_EXT'] = df[ext_cols].sum(axis=1)
df['sum_EXT']
```

```python
# Move 'sum_EXT' after 'CON10'
con10_index = df.columns.get_loc('CON10')
cols = list(df.columns)
cols
```

```python
# Move 'sum_EXT' into the right place
cols.insert(con10_index + 1, cols.pop(cols.index('sum_EXT')))
df = df[cols]
df
```

```python
# Write a file
df.to_csv('big_five_dataset_08_with_sum_EXT.csv', index=False)
```
