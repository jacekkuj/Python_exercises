```python
# Define gpt_response_times list
gpt_response_times = [3.74, 29.977, 9.401, 34.669, 1.117, 6.76, 29.259, 33.058, 10.98, 21.921]
```

```python
# Calculation of the arithmetic mean
mean = sum(gpt_response_times) / len(gpt_response_times)
print("Average ChatGPT response time:", mean)
```

```python
# Calculation of median
n = len(gpt_response_times)
index = n // 2

# Check if number of elements is odd or even.
# If there is the remainder of dividing the number of elements in the list, median is value at the index.
# Otherwise, we need to get two middle values from list and calculate arithmetic mean

if n % 2:
    median = sorted(gpt_response_times)[index]
else:
    median = sum(sorted(gpt_response_times)[index - 1:index + 1]) / 2
print("Median ChatGPT response time:", median)

```

```python
# Calculation of mode
mode = max(set(gpt_response_times), key = gpt_response_times.count)
print("Mode ChatGPT response time:", mode)
```

```python
# Calculation of variance
variance = sum((resp_time - mean) ** 2 for resp_time in gpt_response_times) / (len(gpt_response_times) - 1)
print("Variance ChatGPT response times:", variance)
```

```python
# Calculation of standard deviation
stdev = variance ** 0.5
print("Standard deviation of ChatGPT response times:", stdev)
```

```python
import numpy as np

# Dane o cenach nieruchomości
real_estate_prices = {
    'Broadway': 1200000,
    'Park Avenue': 3500000,
    'Madison Avenue': 2800000,
    'Fifth Avenue': 5000000,
    'Lexington Avenue': 2100000,
    'West End Avenue': 1800000,
    'Central Park West': 4500000,
    'Riverside Drive': 1600000,
    'Waverly Place': 1100000,
    'Bleecker Street': 900000,
    'Christopher Street': 950000,
    'Houston Street': 1300000,
    'Canal Street': 800000,
    'Bowery': 700000,
    'Delancey Street': 1000000
}

# Wyodrębnienie cen nieruchomości i przechowanie ich w tablicy NumPy
prices = np.array(list(real_estate_prices.values()))

# Obliczenia
average_price = np.mean(prices)
median_price = np.median(prices)
min_price = np.min(prices)
max_price = np.max(prices)
variance_price = np.var(prices)
std_deviation_price = np.std(prices)

# Wyświetlenie wyników
print(f'Średnia cena nieruchomości: {average_price}')
print(f'Mediana ceny nieruchomości: {median_price}')
print(f'Wartość minimalna ceny nieruchomości: {min_price}')
print(f'Wartość maksymalna ceny nieruchomości: {max_price}')
print(f'Wariancja ceny nieruchomości: {variance_price}')
print(f'Odchylenie standardowe ceny nieruchomości: {std_deviation_price}')
```

```python
from sklearn.datasets import load_wine
import pandas as pd

wine_data = load_wine()
wine_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)
wine_df
```

```python
# Obliczenie średnich wartości
mean_values = wine_df.mean()

# Obliczenie mediany
median_values = wine_df.median()

# Wyświetlenie wyników
print("Średnie wartości:\n", mean_values)
print("\nMediana wartości:\n", median_values)
```

```python
# Dla zmiennej alcohol
min_alcohol = wine_df['alcohol'].min()
max_alcohol = wine_df['alcohol'].max()
std_alcohol = wine_df['alcohol'].std()

# Dla zmiennej malic_acid
min_malic_acid = wine_df['malic_acid'].min()
max_malic_acid = wine_df['malic_acid'].max()
std_malic_acid = wine_df['malic_acid'].std()

# Wyświetlenie wyników
print(f"\nZmienne 'alcohol': Min: {min_alcohol}, Max: {max_alcohol}, Std: {std_alcohol}")
print(f"Zmienne 'malic_acid': Min: {min_malic_acid}, Max: {max_malic_acid}, Std: {std_malic_acid}")
```

```python
# Dla zmiennej total_phenols
q1_total_phenols = wine_df['total_phenols'].quantile(0.25)
q3_total_phenols = wine_df['total_phenols'].quantile(0.75)

# Dla zmiennej proline
q1_proline = wine_df['proline'].quantile(0.25)
q3_proline = wine_df['proline'].quantile(0.75)

# Wyświetlenie wyników
print(f"\nZmienne 'total_phenols': Q1: {q1_total_phenols}, Q3: {q3_total_phenols}")
print(f"Zmienne 'proline': Q1: {q1_proline}, Q3: {q3_proline}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obliczenie macierzy korelacji
correlation_matrix = wine_df.corr()

# Wizualizacja macierzy korelacji za pomocą seaborn
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt='.2f')
plt.title('Macierz korelacji zmiennych liczbowych ciągłych')
plt.show()
```
